```python
import requests, re, time, json
from bs4 import BeautifulSoup
```

```python
url = 'https://www.lettuceclub.net/recipe/genre/1/ninki/p1/'
```

```python
headers = {
    'User-Agent': "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.90 Safari/537.36"
}
response = requests.request("GET", url, headers=headers)
```

```python
def get_recipe_urls(url):
    headers = {
        'User-Agent': "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.90 Safari/537.36"
    }
    response = requests.request("GET", url, headers=headers)
    data = BeautifulSoup(response.text, 'html.parser')
    pattern = re.compile(r'^\/i\/R1\/img\/dish\/\d+\/[^\/]+\.(jpg|png)\?.*$')
    links = data.find_all('a', href=True)
    recipe_urls = []
    for link in links:
        if '/recipe/dish/' in link['href']:
            imgs = link.find_all('img')
            recipe_img = [img for img in imgs if pattern.match(img['src'])]
            # print(link['href'], len(recipe_img), recipe_img)
            recipe_urls.append({
                'url': f"https://www.lettuceclub.net{link['href']}",
                'image': f"https://www.lettuceclub.net{recipe_img[0]['src']}" if recipe_img else None,
                'title': recipe_img[0]['alt'] if recipe_img else None,
            })
    return recipe_urls
```

```python
res = get_recipe_urls(url)
res
```

```python
genres = [
    '',
    '和食',
    '洋食',
    '中華',
    'イタリアン',
    'エスニック',
    '韓国風',
    'その他',
]
```

```python
recipe_urls = {}
for genre_num in range(1,8):
    # print(genre)
    genre_name = genres[genre_num]
    recipe_urls[genre_name] = []
    for page_num in range(1,6):
        # print(page_num)
        url = f'https://www.lettuceclub.net/recipe/genre/{genre_num}/ninki/p{page_num}/'
        print(url)
        urls = get_recipe_urls(url)
        time.sleep(2.)
        recipe_urls[genre_name] += urls
    # break
```
