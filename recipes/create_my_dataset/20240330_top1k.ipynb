```python
import json
with open('../../data/foodcom_recipes.jsonl', 'r') as file:
    recipes_data = [json.loads(line) for line in file]
```

```python
# レシピデータからrating.averageとrating.nが大きいTop-1000を抽出するコード
top1000_recipes = [recipe for recipe in recipes_data if recipe['top_1000']]
```

```python
# top1000_recipes[0]
```

```python
unique_ingredients = set()
for recipe in top1000_recipes:
    for ingredient in recipe['ingredients']['names']:
        unique_ingredients.add(ingredient)

# ユニークな成分のリストを表示
unique_ingredients_list = list(unique_ingredients)
print(len(unique_ingredients_list))
print(unique_ingredients_list)
```

```python
unique_tags = set()
for recipe in top1000_recipes:
    for tag in recipe['tags']:
        unique_tags.add(tag)

# ユニークな成分のリストを表示
unique_tags_list = list(unique_tags)
print(len(unique_tags_list))
print(unique_tags_list)
```
